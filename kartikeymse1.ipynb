```python
import pandas as pd

try:
  df = pd.read_csv('healthcare_data (1).csv')
  display(df.head())
  print(df.shape)
except FileNotFoundError:
  print("Error: File not found.  Please ensure the file 'healthcare_data (1).csv' is in the current working directory.")
except Exception as e:
  print(f"An error occurred: {e}")
```

```python
# Examine the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Check data types
print("\nData Types:\n", df.dtypes)

# Descriptive statistics for numerical features
print("\nDescriptive Statistics:\n", df.describe())

# Identify missing values
print("\nMissing Values:\n", df.isnull().sum())

# Explore unique values and distributions of categorical features
for col in df.columns:
    if df[col].dtype == 'object':  # Check for categorical columns (object dtype)
        print(f"\nUnique values in '{col}':\n{df[col].unique()}")
        print(f"\nValue counts in '{col}':\n{df[col].value_counts()}")
    elif df[col].nunique() < 10: # Check for numerical columns that might be categorical
        print(f"\nUnique values in '{col}':\n{df[col].unique()}")
        print(f"\nValue counts in '{col}':\n{df[col].value_counts()}")
```

```python
import matplotlib.pyplot as plt

# Numerical features to check for outliers
numerical_features = ['Age', 'BloodPressure', 'SugarLevel', 'Weight']

# Create box plots to visualize outliers
plt.figure(figsize=(12, 8))  # Adjust figure size for better visualization
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 2, i + 1)
    plt.boxplot(df[feature])
    plt.title(f'Box Plot of {feature}')
    plt.ylabel(feature)
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()

# Calculate IQR and identify outlier boundaries for each feature
for feature in numerical_features:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Winsorize outliers (capping at 5th and 95th percentiles)
    df[feature] = df[feature].clip(lower=df[feature].quantile(0.05), upper=df[feature].quantile(0.95))
    print(f"Winsorized {feature}")


# Create box plots after outlier handling
plt.figure(figsize=(12, 8))  # Adjust figure size for better visualization
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 2, i + 1)
    plt.boxplot(df[feature])
    plt.title(f'Box Plot of {feature} (After Winsorizing)')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Descriptive Statistics
numerical_features = ['Age', 'BloodPressure', 'SugarLevel', 'Weight']
print("Descriptive Statistics:\n", df[numerical_features].describe())

# 2. Correlation Analysis
correlation_matrix = df[numerical_features].corr(method='pearson')
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# 3. Distribution Analysis
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```
